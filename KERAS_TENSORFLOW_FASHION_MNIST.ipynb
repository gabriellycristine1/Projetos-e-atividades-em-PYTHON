IMPORTANDO AS BIBLIOTECAS

```python
import tensorflow 
from tensorflow import keras
import matplotlib.pyplot as plt
```

IMPORTANDO DATASET QUE O PROPRIO KERAS DISPONIBILIZA

```python
Dataset = keras.datasets.fashion_mnist
Dataset.load_data()
```

SEPARANDO ENTRE TREINO E TESTE

```python
((X_imagens_treino, Y_identificacoes_treino),(X_imagens_teste, y_identificacoes_teste)) = Dataset.load_data()
```

```python
modelo = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                           keras.layers.Dense(256, activation=tensorflow.nn.relu)

]  

)
```

```python
modelo.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy')
modelo.fit(X_imagens_treino, Y_identificacoes_treino)
```

TREINANDO E COMPILANDO O MODELO

```python
X_imagens_treino = X_imagens_treino/float(255)
modelo = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                           keras.layers.Dense(256, activation=tensorflow.nn.relu),
                           keras.layers.Dense(10, activation=tensorflow.nn.softmax)

]  

)
modelo.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy',metrics='accuracy')
modelo.fit(X_imagens_treino, Y_identificacoes_treino, epochs=5)
```

```python
import numpy as np
```

TESTANDO MODELO

```python
teste = modelo.predict([X_imagens_teste])
teste[0]

print('resultado Teste:',np.argmax(teste[0]))
print('numero da imagem de teste', y_identificacoes_teste[0])
```

VALIDANDO MODELO COM PRECISÃO FINAL DE 97,72%

```python
historico = modelo.fit(X_imagens_treino, Y_identificacoes_treino, epochs=30, validation_split=0.2)
```

```python
historico.history
```

CRIANDO GRAFICO PARA ANALISAR O COMPORTAMENTO DA IA DURANTE OS TESTES FEITOS

```python
plt.plot(historico.history['accuracy'])
```

```python
plt.plot(historico.history['loss'])
plt.plot(historico.history['val_loss'])
plt.title('PERDAS POR EPOCAS')
plt.xlabel('EPOCAS')
plt.ylabel('PERDAS')
plt.legend(['treino','validação'])
```
